### Analysis

- For mice in the Capomulin treatment group, body weight is positively associated with tumor volume.
- Median tumor volume was lowest for the Ramicane treatment group.
- The Ceftamin treatment group had the largest IQR for tumor sizes and therefore more variation in treatment outcomes relative to the other three groups.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from pathlib import Path

# Study data files
mouse_metadata_to_load = Path("data/Mouse_metadata.csv")
study_results_to_load = Path("data/Study_results.csv")

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_to_load)
study_results = pd.read_csv(study_results_to_load)

# Combine the data into a single DataFrame
scc_data_complete_df = pd.merge(mouse_metadata, study_results, how="left", on=['Mouse ID', 'Mouse ID'])

# Display the data table for preview
scc_data_complete_df.head()
```

```python
# Check the number of mice.
total_unique_mice = scc_data_complete_df['Mouse ID'].nunique()
print(f'Total number of unique mice:{total_unique_mice}')
```

```python
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint
duplicates = scc_data_complete_df.duplicated(subset=['Mouse ID', 'Timepoint'])

# Print the duplicate records
print("Duplicate records:")
print(scc_data_complete_df[duplicates])
```

```python
# Filter out records with Mouse ID 'g989'
filtered_df = scc_data_complete_df[scc_data_complete_df['Mouse ID'] != 'g989']

# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint
duplicates = filtered_df.duplicated(subset=['Mouse ID', 'Timepoint'])

# Use .loc to select the rows that are not duplicates
clean_df = filtered_df.loc[~duplicates]
```

```python
# Check the unique values of 'Mouse ID' in the cleaned DataFrame
unique_mice = clean_df['Mouse ID'].nunique()
print("Total unique mice after removing duplicates:", unique_mice)
```

## Quartiles, Outliers and Boxplots

```python
unique_timepoints = clean_df['Timepoint'].unique()
print(unique_timepoints)
```

```python
# Filter data for the specified treatment regimens
drug_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
filtered_df = clean_df.loc[clean_df["Drug Regimen"].isin(drug_list)]

# Find the final time and final tumor volume for each mouse
final_time = filtered_df.groupby('Mouse ID')['Timepoint'].max()
final_vol_df = pd.DataFrame(final_time)

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
filtered_merge = pd.merge(final_vol_df, clean_df, on=['Mouse ID', 'Timepoint'], how='left')

# Display the final tumor volume of each mouse
filtered_merge.head()
```

```python
# Calculate the quartiles and IQR, and determine if there are any potential outliers across all four treatment regimens.
# Create an empty list to hold the tumor volume data.
tumor_vol_list = []

# Loop through each drug in the treatment list, locating the rows in the merged DataFrame that correspond to each treatment. 
for treatment in drug_list:
    treatment_data = filtered_merge.loc[filtered_merge["Drug Regimen"] == treatment,'Tumor Volume (mm3)']

    # Calculate quartiles and IQR
    quartiles = treatment_data.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq

    # Find outliers
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outliers = treatment_data[(treatment_data < lower_bound) | (treatment_data > upper_bound)]

    # Print out a summary with potential outliers identified for each treatment
    print(f"Summary statistics for {treatment}:")
    print(f"The lower quartile is: {lowerq}")
    print(f"The upper quartile is: {upperq}")
    print(f"The interquartile range is: {iqr}")
    print(f"The the median is: {quartiles[0.5]} ")
    print(f"Potential outliers: {outliers}\n")
    
# Append the resulting final tumor volumes for each drug to the empty list.
tumor_vol_list.append(treatment_data)
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.

#Create an empty list to store tumor volume for each drug
treatment_data_list = []

# Iterate through each drug in the drug list
for treatment in drug_list:
    treatment_data = filtered_merge.loc[filtered_merge["Drug Regimen"] == treatment,'Tumor Volume (mm3)']
    treatment_data_list.append(treatment_data)
    
# Create a figure and establish y axis   
fig, ax = plt.subplots()
ax.set_title('Tumor Volume for each Treatment Group')
ax.set_ylabel('Tumor Volume (mm3)')

# Plot all four boxplots together
ax.boxplot(treatment_data_list, labels=drug_list)
plt.show()
```

```python
# Plot the linear regression model on top of the previous scatter plot.

import statsmodels.api as sm
capomulin_mice_avg_vol = filtered_merge[filtered_merge["Drug Regimen"] == 'Capomulin'].groupby('Mouse ID')['Tumor Volume (mm3)'].mean()

plt.scatter(filtered_merge[filtered_merge['Drug Regimen'] == 'Capomulin']['Weight (g)'], capomulin_mice_avg_vol)
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Mouse Weight vs Avg Tumor Volume for Capolmulin Treatment')

# Regress "weight" onto "tumor volume" (plus an intercept)
X = np.column_stack((np.ones_like(capomulin_mice_avg_vol), filtered_merge[filtered_merge['Drug Regimen'] == 'Capomulin']['Weight (g)']))
model = sm.OLS(capomulin_mice_avg_vol, X).fit()

# Plot regression line
plt.plot(X[:, 1], model.predict(X), color='red', linewidth=2)
plt.show()
```

```python
# Calculate the correlation coefficient and a linear regression model for mouse weight and observed tumor volume for the entire Capomulin regimen
import statsmodels.formula.api as sm

capomulin_mice = capomulin_mice.rename(columns={'Weight': 'Weight (g)', 'Tumor Volume': 'Tumor Volume (mm3)'})

# Fitting the model
model = sm.ols(formula='Q("Weight (g)") ~ Q("Tumor Volume (mm3)")', data=capomulin_mice).fit()

# Model summary
print(model.summary())
```
